```python
#PROGRAMMING QUESTION
#Q1
items=[]
for i in range(1000,3000):
    s=str(i)
    if (int(s[0])%2==0) and (int(s[1])%2==0) and (int(s[2])%2==0):
        items.append(s)
print(",".join(items))
```
